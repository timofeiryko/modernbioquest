# Select the data to upload

```python
part_1_df = df[(df.grade == 11) & (df.stage == 'РЭ') & (df.part == 1)].sort_values(['year', 'number'])
part_1_df
```

```python
part_2_df = df[(df.grade == 11) & (df.stage == 'РЭ') & (df.part == 2)].sort_values(['year', 'number'])
part_2_df
```

```python
part_1_df = part_1_df[part_1_df.year != 2022]
```

```python
# Two cells below are not DRY, but I'm lazy to reuse the code from pdfparsing
```

```python
@dataclass
class QuestionDraft:
    title: str
    text: str
    part: int
    number: int

@dataclass
class CleanedQuestion:
    question: QuestionDraft
    text: str
    answer_variants: List[str]
```

```python
cleaned_questions = {}
for i, row in pd.concat([part_1_df, part_2_df], axis=0).iterrows():

    question = QuestionDraft(
        title='',
        text=row.text,
        part=row.part,
        number=row.number
        
    )

    splitted = re.split('(\s[а-яА-Я1-9]\)\s[а-яА-Я1-9a-zA-Z])', question.text)
    cleaned_text = ' '.join(splitted[0].split())

    dividers = splitted[1::2]
    variants = splitted[2::2]
    variants = [' '.join(re.split('\s', (v + d))) for v, d in zip(dividers, variants)]
    
    cleaned_questions[i] = CleanedQuestion(question, cleaned_text, variants)
```

```python
OUTPUT_PATH = 'reg_11_2018-2021_parts_1-2.p'
```

```python
pickle.dump(cleaned_questions, open(OUTPUT_PATH, 'wb'))
```
